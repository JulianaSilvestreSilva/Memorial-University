### Introduction
Cinema has been a part of the entertainment industry for a long time. It creates a massive impact on people all over the world and it allows viewers to learn, reflect and have fun.

The history of cinema begins with an impressive technological phenomenon in our history, the ability to capture the “image-movement”, that is, the capture of dynamic images of reality, and not static ones, as is the case with photography. The capture of the “motion-image” was made possible from 1889 onwards with the creation of the kinetoscope by William Dickson, assistant to the American scientist and inventor Thomas Edison. This invention and the models that followed it in the following decade contributed to the development of cinema as we understand it today, that is, cinematographic art.

over the years, with the technological development, the films evolved and, for example, started to have colors and a longer duration. ### The dataset

Then, in order to determine whether the variables are symmetrically distributed, the histograms below were plotted. In addition, bloxpots were used to identify possible outliers present in the database.

```python
# finding out if the range variables are normally distributed by making a histogram with quantiles and a density plot
def plot_histograms_density(df, columns):
    fig, axs = plt.subplots(len(columns),1,figsize=(16,18))
    plt.subplots_adjust(left=0.1, 
                    bottom=0.1,  
                    right=0.9,  
                    top=0.9,  
                    wspace=0.4,  
                    hspace=0.4) 
    i = 0
    for c in columns:
        df[c].hist(ax=axs[i], density=True) # normalizes the density
        df[c].plot.density(ax=axs[i], title=c)
        i+=1

plot_histograms_density(df_range, range_var)
```

```python
# making a boxplot to see if there are outliers in the range variables
def plot_boxplot(copy_df, columns):
    fig, axs = plt.subplots(2,3,figsize=(16,8))
    axs = axs.flatten()
    plt.subplots_adjust(left=0.1, 
                    bottom=0.1,  
                    right=0.9,  
                    top=0.9,  
                    wspace=0.4,  
                    hspace=0.4) 
    i = 0
    for c in columns:
        axs[i].boxplot(copy_df[c])
        axs[i].set(title = c)
        i+=1

plot_boxplot(df_range, range_var)
```

From the analysis of the plots above, we can conclude that the distribution of such data is skewed. In this way we looked for the best way to remove outliers without removing too many instances.

Initially, the Log transformation was performed, in order to transforms our values to onto a scale that approximates normality, and for that, we present the skew of the variables before and after the transformation, as well as the new histograms and bloxpots. 

```python
df_range_2 = df_range.copy()
# displaying the skew of the range variables before the transformation
print("skew of the range variables before the transformation:")
df_range.skew()
```

```python
# applying a log transformation to the range variables
df_range_2['duration'] = np.log(df_range['duration'])
df_range_2['gross'] = np.log(df_range['gross'])
df_range_2['num_voted_users'] = np.log(df_range['num_voted_users'])
df_range_2['budget'] = np.log(df_range['budget'])
df_range_2['imdb_score'] = np.log(df_range['imdb_score'])
df_range_2['title_year'] = np.log(df_range['title_year'])

# displaying the skew of the range variables after the log transformation
df_range_2.skew()
```

```python
# making a histogram with quantiles and a density plot to see if the range variables are normally distributed after the log transformation
plot_histograms_density(df_range_2, range_var)
```

```python
# making a boxplot to see if there are outliers in the range variables after the log transformation
plot_boxplot(df_range_2, range_var)
```

From the boxplots generated above, we can conclude that the log transformation proved to be useful in reducing data dispersion, however outliers remain present in the database.

In this way, we chose to use an approach that identifies outliers as values that are 3 times greater than the standard deviation. Therefore, the method below calculates for each column the min and max values within this range and replaces it by the max value.

```python
# Removing outliers by using standard deviation method
# Storing the values in a new dataframe
def replace_outliers_std(df, columns):
    for c in columns:
        mean = df[c].mean()
        std = df[c].std()
        low_lim = mean - 3*std
        high_lim = mean + 3*std
        print(f"High ({c}):", high_lim)
        print(f"Low ({c}):", low_lim)
        df[c] = np.where(df[c] >= high_lim, high_lim, np.where(df[c] <= low_lim, low_lim, df[c]))

df_range_std = df_range_2.copy()
replace_outliers_std(df_range_std, range_var)
```

```python
# showing the skew of the range variables after using standard deviation method
df_range_std.skew()
```

```python
# showing the boxplot of the range variables after removing the outliers
def plot_boxplot_wh3 (copy_df, columns):
    fig, axs = plt.subplots(2,3,figsize=(16,8))
    axs = axs.flatten()
    plt.subplots_adjust(left=0.1, 
                    bottom=0.1,  
                    right=0.9,  
                    top=0.9,  
                    wspace=0.4,  
                    hspace=0.4) 
    i = 0
    for c in columns:
        axs[i].boxplot(copy_df[c], whis=3)
        axs[i].set(title = c)
        i+=1

plot_boxplot_wh3(df_range_std, range_var)
```

This technique therefore proved to be adequate to remove outliers. Thus, we have the database clean and we can move on to the next step, which is exploring data scaling techniques.

### Data scaling pre-assessment

The data scaling step involves transforming the data so that it fits within a specific scale. Therefore, improving the performance of some machine learning algorithms. For this, the results obtained by different normalization and standardization techniques were analyzed.

Initially, in order to analyze the best data scaling techniques, all transformations were plotted side by side as follows. Min-Max scaling, Max-abs scaler, Robust scaler and Quantile transform scaler are normalization techniques. And, on the other hand, Z-Score and Log are standardization techniques.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import FunctionTransformer

# Showing all Scalers effect on data 
def plot_histograms_density_for_scalers(df, columns, fig_title, scalers, scaler_names):
    fig, axs = plt.subplots(len(columns),len(scalers)+1,figsize=(16,18),constrained_layout=True)
    fig.suptitle(fig_title, fontsize=16)
    axs = axs.flatten()
    i = 0 
    for c in columns:
        df[c].hist(ax=axs[i], density=True) # normalizes the density
        df[c].plot.density(ax=axs[i], title=c)
        i+=1
        for j in range(len(scalers)):
            df_transformed = scalers[j].fit_transform(df)
            df_transformed = pd.DataFrame(df_transformed, index=df.index, columns=df.columns)
            df_transformed[c].hist(ax=axs[i], density=True, stacked=True) # normalizes the density
            df_transformed[c].plot.density(ax=axs[i], title=scaler_names[j])
            i+=1
            

scaler_names = ['MinMax', 'Z-score', 'MaxAbs', 'Robust', 'Quantile', 'Log']
arr_scalers = [MinMaxScaler(), Normalizer(), MaxAbsScaler(), RobustScaler(), QuantileTransformer(n_quantiles=10, random_state=0, output_distribution='normal'),FunctionTransformer(np.log1p)]
plot_histograms_density_for_scalers(df_range_std, range_var, 'All Scalers', arr_scalers, scaler_names)
```

From the analysis of the results obtained above, it can be seen that the best result was using the normalization techniques, since the data is not symmetrically distributed, as it is possible to observe by the initially plotted histograms. In this way, we can observe that among the normalization techniques, the Quantile transform scaler obtained the worst result, since it mapped the variable’s probability distribution to another probability distribution. In this way, for a better visualization of the results obtained with MinMax, MaxAbs and Robust, such approaches have been applied again below.

```python
# Before and after MinMaxScaler transformations

import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

def plot_histograms_density(df, columns, df_transformed, fig_title):
    fig, axs = plt.subplots(len(columns),2,figsize=(16,18), constrained_layout=True)
    fig.suptitle(fig_title, fontsize=16)
    axs= axs.flatten()
    plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
    i = 0
    for c in columns:
        df[c].hist(ax=axs[i], density=True)
        df[c].plot.density(ax=axs[i], title=c)
        df_transformed[c].hist(ax=axs[i+1], density=True, stacked = True)
        df_transformed[c].plot.density(ax=axs[i+1], title=str(c + '_transformed'))
        i+=2
```

As a baseline strategy, it was chosen to replace the missing values using a centrality measure to replace the values. In this way, we first retrieve the database before excluding the missing attributes and apply the method that replaces such values by the median.

With regard to outliers, the base strategy was the one demonstrated in the first part of iteration two of the present work, which corresponds to identifying outliers as values that are 3 times greater than the standard deviation and calculate for each column the min and max values within this range and replace it by the max value, as it was done in the function replace_outliers_std.

```python
def plot_series_side_by_side(original, resulting, columns):
    fig, axs = plt.subplots(len(columns),1,figsize=(16,18))
    axs = axs.flatten()
    custom_xlim = (original.index.min(), original.index.max())
    i = 0
    for c in columns:
        resulting[c].plot(ax=axs[i], title=c, xlabel='', xlim=custom_xlim, label='Modified', lw=4, color='red')
        original[c].plot(ax=axs[i], title=c, xlabel='', xlim=custom_xlim, label='Original', color='black')
        axs[i].legend()
        i+=1
        
fill_agg_df = drop_df.copy()
def transform_aggregate(df, agg_func, columns):
    for c in columns:
        X = None
        if    agg_func == 'median':
            X = df[c].median()
        elif  agg_func == 'mean':
            X = df[c].mean()
        else: 
            X = df[c].mode()[0]
        df[c].fillna(X, inplace=True) 

transform_aggregate(fill_agg_df, 'mean', ['duration', 'gross', 'num_voted_users', 'budget', 'title_year', 'imdb_score'])
# transform_aggregate(fill_agg_df, 'median', ['nox_values','no_values','no2_values','pm25_values','pm10_values'])
# transform_aggregate(fill_agg_df, 'mode', val_cols_names)
plot_series_side_by_side(drop_df, fill_agg_df, range_var)
```

```python
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import FunctionTransformer
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score

# Using Linear Regression to predict the values from title_year
scaler_names = ['MinMax', 'Z-score', 'MaxAbs', 'Robust', 'Quantile']
arr_scalers = [MinMaxScaler(), Normalizer(), MaxAbsScaler(), RobustScaler(), QuantileTransformer(n_quantiles=10, random_state=0, output_distribution='normal')]
results = {}
model = LinearRegression()

for i in range (len(arr_scalers)):
    # X is all the other variables
    X = df_range_std.drop(['title_year'], axis=1).to_numpy()
    X_t = arr_scalers[i].fit_transform(X)
    # y is NOX that we want to forecast
    y = df_range_std['title_year'].to_numpy()
    # Let's estimate the values for several rounds of crossvalidation
    cv = RepeatedKFold(n_splits=10, n_repeats=5, random_state=1)
    scores = cross_val_score(model, X_t, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
    results[scaler_names[i]] = [abs(elem) for elem in scores]


cv = RepeatedKFold(n_splits=10, n_repeats=5, random_state=1)
scores = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
results['Raw data'] = [abs(elem) for elem in scores]

df_res = pd.DataFrame(results)
boxplot = df_res.boxplot() 
boxplot.set_ylabel('MSE')
boxplot.set_title('Linear Regression')
plt.show()
```
